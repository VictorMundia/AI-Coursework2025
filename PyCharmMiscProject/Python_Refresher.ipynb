```python
days = int(input("Enter the number of days: ")) #Number of days

seconds = days*24*60*60 #conversion of days to seconds

print(f"{days} days = {seconds} seconds") # Output
```

```python
import math

radius = float(input("Enter the radius: ")) #Input the radius

volume = (4/3)*math.pi*(radius**3) # Use the formula to get the volume

print(f"The volume of the sphere is {volume}") #Output
```

```python
#Function to get the area
def area(side):
    return side*side

#Function to get the perimeter
def perimeter(side):
    return side * 4

side = float(input("Enter the length of one side of the square: "))

#Output
print(f"The perimeter of the square is {perimeter(side)}")
print(f"The area of the square is {area(side)}")
```

```python
# Create function to check bot cases
def check_case(ch):
    if ch.isupper():
        return "Uppercase"
    elif ch.islower():
        return "Lowercase"
    else:
        return "Not a letter"

char = input("Enter a character: ")

#Output
print(check_case(char))
```

```python
# Initialize values
x = 0
y = 20

# Repeat until y is less than 6
while y >= 6:
    y = y - 4
    x = x + (2 / y)

# Output
print("x =", x)
```

```python
# Create an Empty list (array)
values = []

# a loop to get 5 values
for i in range(5):
    num = float(input(f"Enter value {i+1}: "))
    values.append(num)

average = sum(values) / len(values)

#Output
print("The average is:", average)
```
